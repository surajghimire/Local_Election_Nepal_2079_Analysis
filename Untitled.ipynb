```python
import requests
import time
```

```python


ALL_PROVINCE_URL="http://result.election.gov.np/JSONFiles/JSONMap/geojson/Province.json"
STATE_URL="http://result.election.gov.np/JSONFiles/JSONMap/geojson/District/STATE_C_{}.json"
DISTRICT_URL="http://result.election.gov.np/JSONFiles/JSONMap/geojson/LL/DCODE_{}.json"
PALIKA_URL="http://result.election.gov.np/JSONFiles/Election2079/Local/{}.json"


for state_num in range(1,8):
	print ('\nfor state ', state_num)
	districts = requests.get(STATE_URL.format(state_num)).json()['features']
	for district in districts:
		print (district['properties']['DISTRICT'])
		dist_code = district['properties']['DCODE']
		palikas = requests.get(DISTRICT_URL.format(dist_code)).json()['features']
		for palika in palikas:
			palika_id = palika['properties']['GN_CODE']
			palika_name = palika['properties']['GaPa_NaPa']
			print (palika['properties']['GaPa_NaPa_'])
			if palika_id == 5691: # remove this line to scrape all palikas 
			#if true:
				time.sleep(5)  # to avoid rate limiting
				with open("{}_{}_{}_{}.json".format(state_num, dist_code, palika_id, palika_name), "w") as f:
					f.write(requests.get(PALIKA_URL.format(palika_id)).text)
```
